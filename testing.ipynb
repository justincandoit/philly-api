```python
import requests
import json
```

```python
base = "https://phl.carto.com/api/v2/sql?q="

query = "SELECT * from opa_properties_public where "

params = """
            geographic_ward = '31' 
            and location='2053 AMBER ST'
            and frontage > 17
            -- and sale_date like '192%'
            
            limit 10
            

            """
```

```python
url = base+query+params
response = requests.get(url).content
response_str = response.decode('utf-8')
print(url)
```
